# Formato Condicional
## Definicion
> Es una herramienta presente en la mayoría de hojas de cálculo (como Microsoft Excel®), permite modificar el formato de una celda o rango de estas, de acuerdo al cumplimiento o no de ciertas condiciones. Por ejemplo, cambiar el color de relleno de una celda si su valor no sobrepasa cierto límite
>
>Se entiende por formato todas las características visuales e identificables de una celda, muchas de las herramientas usadas para modificar o aplicar el formato son accesibles a través de:
>
>        Ficha Inicio > Grupo: fuente
>
>
