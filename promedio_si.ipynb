# Funcion PROMEDIO.SI
## Concepto
> La funcion promedio.si hace parte del grupo de _funciones condicionadas_, donde su caracteristica principal es el uso de un campo sujeto a una condicion, la mayoria de veces estas se resuelven usando los operadores relacionales:
>
> | Operador       | Signo | Definicion |
|----------------|-------|------------|
| Mayor que      | >     | A > B      |
| Menor que      | <     | A < B      |
| Igual a        | =     | A = B      |
| No es igual a  | <>    | A <> B     |
>
> Aclaracion: el uso de la sintaxis de estos operadores es verificada en Excel, puede variar en otros software o lenguaje de programacion.
## Estructura
> Consta de tres campos:
> ### Rango
>> Corresponde a la columna de datos que contienen la condicion para ser relacionada en el calculo del promedio.
> ### Criterio
>> Es la condicion a evaluar, dentro de los datos de la columna _Rango_.
> ### Rango promedio
>> Corresponde a la columna de datos donde se desea calcular el promedio. Generalmente es diferente a la columna rango, puede darse el caso en donde los tres campos pertenecen a la misma columna.

## Analisis del Problema
> ### Identificando los tres campos
>> AL leer el enunciado se tiene claro lo siguiente:
>> - El promedio a calcular corresponde a la distancia en Kilometros - Columna B y corresponderia al campo __Rango promedio__.
>> - Ahora se debe identificar la columna donde se encuentra la condicion a la cual estan sujetos los datos, segun el enunciado se debe tener en cuenta solamente las ciudades con salida al mar, la columna que proporciona esta informacion es la D, por lo tanto seria el campo __Rango__.
>> - Por ultimo, se define la condicion, recordemos que esta debe pertenecer a la misma columna del campo _Rango_, interesan unicamente las ciudades con salida al mar, dicha informacion se indica en la tabla con la palabra _Si_, al ser un texto el operador a usar es _igual a_ y la sintaxis final es: ="Si". Dicha expresion corresponde al __Criterio__.
>>
>> __Nota:__ Tenga en cuenta que en caso de ser texto, la expresion en el campo _Criterio_ debe ir entre comillas y estar escrita exactamente igual a como aparece en la tabla, ya que Excel es Case Sensitive.
